### Olympic 2024 Data Merge

This script merges two datasets: 

* "Olympic2024_Medallists.csv" - A dataset containing the medal winners.
* "Olympic2024_Athletes.csv" - A dataset with general athlete information.

Objective:
By merging these datasets, we can enrich the medalist data with additional athlete details such as age, country, and sport.
This allows for deeper analysis, such as understanding the age distribution of medalists or comparing performance by nationality.

Merge Key:
Both datasets share a common key: **'Athlete Code'**, which will be used to combine the information.


### Lets import libraries and load both the datasets

```python
import pandas as pd
```

```python
# Load the first dataset (Medallists)

medallists_data = pd.read_csv("../data/Olympic2024_Medallists.csv")
```

```python
medallists_data.head(10)
```

```python
# Load the first dataset (Athletes)

athletes_data = pd.read_csv("../data/Olympic2024_Athletes.csv")
```

```python
athletes_data.head(10)
```

```python
print("Medallists Data Columns:", medallists_data.columns)
print("Athletes Data Columns:", athletes_data.columns)
```

Since the dataset has different column header we need to rename the other one

```python
# Rename the column in athletes_data to match medallists_data
athletes_data.rename(columns={"code": "code_athlete"}, inplace=True)
```

```python
# Now merge two datasets

merged_datasets = pd.merge(medallists_data, athletes_data, on="code_athlete", how="inner")
```

```python
print("Merge successful! Merged dataset has", merged_datasets.shape[0], "rows and", merged_datasets.shape[1], "columns.")
```
